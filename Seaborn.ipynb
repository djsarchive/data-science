```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df2=pd.DataFrame(np.random.randint(1,100,(4,3)),index=["Devshree","Mariyam","Fardin","Afzal"],columns=["DSA","Php",".net"])
df2
```

```python
x =df2.index
y =df2["Php"]
plt.subplot(1,2,1)
plt.plot(x, y,"g^-")
plt.xlabel("marks")
plt.ylabel("students")
plt.title("php marks ")
plt.grid(color='b', linestyle='--' )
plt.ylim(0,100)
plt.legend(['First line', 'Second line'])
plt.show()




x =df2.index
y =df2["DSA"]
plt.subplot(1,2,1)
plt.plot(x, y,"r", marker='*', linestyle='-',data=df2)
plt.xlabel("marks")
plt.ylabel("students")
plt.title("DSA marks ")
plt.grid(color='r', linestyle='--' )
```

```python
#legend
x =df2.index
y =df2["Php"]
plt.plot(x, y,"g^-")
plt.xlabel("student name")
plt.ylabel("student marks")
plt.title(" marks ")
y =df2["DSA"]
plt.plot(x, y,"r*-")
plt.grid(color='r', linestyle='--' )
plt.ylim(0,100)
y =df2[".net"]
plt.plot(x, y,"y*-")

plt.legend(['Php', 'DSA' ,".net"])
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
```

```python
#Jointplot
#parameters in jointplot fpr 'kind' are: scatter, reg, resid, kde, hex
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#pairplot: allows us to plot pairwise relationships between variables within a dataset
sns.pairplot(tips)
```
